**3)Find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of
India.**

```python
#Top 10 cuisines served by maximum number of restaurents in Delhi NCR.
df1=df[df['City']=='Delhi NCR']
cuisines=df1['Cuisines'].str.split(",").sum()
df1=pd.Series(cuisines)
df1=df1.value_counts().reset_index()
df1.rename(columns={"index":"Cuisine",0:'Number of restaurents'},inplace=True)
df1=df1.sort_values('Number of restaurents',ascending=False)[0:10]
df1
```

```python
fig=plt.figure(figsize=(12,6))
sns.barplot(df1,x='Cuisine',y='Number of restaurents')
plt.title("Top 10 Cuisne Served By Most Number of Restaurents in Delhi NCR",fontsize=14,fontstyle='italic')
plt.show()
```

* Top 3 cuisine served by most number of restaurents in delhi NCR are: **North Indian,Chinese,Fast Food**.

```python
#Top 10 cuisines served by maximum number of restaurents in Rest of India.
df1=df[df['City']=='Rest of India']
cuisines=df1['Cuisines'].str.split(",").sum()
df1=pd.Series(cuisines)
df1=df1.value_counts().reset_index()
df1.rename(columns={"index":"Cuisine",0:'Number of restaurents'},inplace=True)
df1=df1.sort_values('Number of restaurents',ascending=False)[0:10]
df1
```

* Top 3 cuisines served by most number of restaurents in rest of india are: **North Indian, Chinese and Continental**.

**5)Compare the price range for restaurants that have table booking vs that have online
booking. Plot suitable graphs.**

```python
#For table booking
df.groupby(['Price range','Has Table booking'])['Has Table booking'].count()
```

```python
#Plotting a count plot
df1=df[(df['Has Table booking']=='Yes')]
fig=plt.figure(figsize=(9,6))
sns.countplot(df1,x='Price range')
plt.title("Price range for Restaurents that have Table Booking",fontsize=14,fontstyle='italic')
plt.ylabel("Number of Restaurents")
plt.show()
```

```python
#For online delivery
df.groupby(['Price range','Has Online delivery'])['Has Online delivery'].count()
```

```python
#Plotting a count plot
df1=df[df['Has Online delivery']=='Yes']
fig=plt.figure(figsize=(9,6))
sns.countplot(df1,x='Price range')
plt.title("Price range for Restaurents that have Online Delivery",fontsize=14,fontstyle='italic')
plt.ylabel("Nunber of Restaurents")
plt.show()
```

* In case of table bookings there are **more number of restaurents present in higher price range as compared to the number of restaurents present in lower price range**,with most number of restaurents present in price range of 3.
* In case of online delivery there are**more number of restaurents present in lower price range as compared to the number of restaurents present in higher price range**,with most number of restaruents present in price range of 2.
* Also we can see that number of restaurent that offer online delivery **are more** as compared to the number of restaurents that offer table bookings which may imply that there is more demand for online delivery as compared to table bookings.
